```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data = pd.read_csv('pais_filtrado.csv')
data.head(5)
```

```python
#Identificar valores nulos por columna
valores_nulos= data.isnull().sum()
valores_nulos
```

```python
#Identificar valores nulos por dataframe
valores_nulos=data.isnull().sum().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data1 = data.copy()
```

```python
#BATHROOMS 
# #Primer método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media
data1["bathrooms"]=data1["bathrooms"].fillna(round(data["bathrooms"].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#BEDROOMS 
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
data1["bedrooms"] =data1["bedrooms"].fillna(method="bfill") 
```

```python
#Corroboramos valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#BEDS
#Sustituir valores nulos por valores no nulos hacia adelante forward fill" ("ffill")
#Filtro por columnas
data1["beds"] =data1["beds"].fillna(method="ffill") 
```

```python
#Corroboramos valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```
